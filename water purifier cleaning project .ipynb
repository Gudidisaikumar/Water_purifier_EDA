## Changing the data type for the columns

```python
ee=[]
for i in df2['Specification']:
    if 'Capacity' in i:
        a=re.findall('\d+\sL/hr',str(i))
        ee.append(a)
    else:
        ee.append(5)
ee
```

```python
capacity=[]
for i in ee:
    if i != 5:
        for j in i :
            j=j.split()
        capacity.append(j[0])
    else:
        capacity.append(5)
capacity
```

```python
for i in range (len(capacity)):
    capacity[i]=int(capacity[i])
```

```python
df2["Capacity (L/hr)"]=capacity
```
